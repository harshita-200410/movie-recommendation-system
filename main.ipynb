Feature Selection Part

```python
movies.columns
```

```python
movies=movies[['id', 'title', 'genre', 'overview']]
```

```python
movies
```

```python
movies['tags'] = movies['overview']+movies['genre']
```

```python
movies
```

```python
new_data = movies.drop(columns=['overview', 'genre'])
```

```python
new_data
```

```python
for feature in movies:
    movies[feature]=movies[feature].fillna('')
```

```python
#combining all the selected features

combined_features = movies['genre'] + ' ' +movies['overview']
```

```python
print(combined_features)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#converting the text data to feature vector

cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vectorizer = CountVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#getting similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
#getting the movie name from user

movies = input('ENTER YOUR FAVOURITE MOVIE NAME : ')
```

```python
#creating a list with all the movies names given in dataset

list_of_all_titles = movies['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the close match for the movie

find_close_match = difflib.get_close_matches(movies, list_of_all_titles)
print (find_close_match)
```

```python

if find_close_match:
    close_match = find_close_match[0]
    print("Close match found:", close_match)
else:
    print("❌ No close match found. ")
```

```python
#find the index of the movie with title 

index_of_the_movie = movies[movies.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting a list of similarity 

similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1], reverse=True)
print(sorted_similar_movies)
```

```python
#print the name of similar movies 

print('MOVIES SUGGESTED FOR YOU:\n')

i = 1
for movie in sorted_similar_movies:
    index = movie[0]  # sorted_similar_movies is list of (index, similarity)
    title_from_index = movies[movies['index'] == index]['title'].values[0]
    
    if i <= 20:  # Show top 30 suggestions
        print(f"{i}. {title_from_index}")
        i += 1
    else:
        break
```

MOVIE RECOMMENDATION SYSTEM

```python
movie_name = input('ENTER YOUR FAVOURITE MOVIE : ')

list_of_all_titles=movies['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies[movies.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1], reverse=True)

print('MOVIES SUGGESTED FOR YOU:\n')

i = 1
for movie in sorted_similar_movies:
    index = movie[0]  # sorted_similar_movies is list of (index, similarity)
    title_from_index = movies[movies['index'] == index]['title'].values[0]
    
    if i <= 20:  # Show top 30 suggestions
        print(f"{i}. {title_from_index}")
        i += 1
    else:
        print("That's all for now! Too many movies, not enough popcorn 🍿😄")
        break
    
```
